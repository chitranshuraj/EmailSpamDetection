```python
#A function letters_only is created; the function filters letters and words from digits and other ASCII Characters
def letters_only(astr):
    return astr.isalpha()
```

```python
#An all_names set is created; all_names set contains all the words imported from nltk.corpus directory
#A WordNetLemmatizer is created; the WordNetLemmatizer reads a word and returns the root form of that word
all_names = set(names.words())
lemmatizer = WordNetLemmatizer()
lemmatizer
```

```python
#A clean_text function is created whose task is to clean the text of any other characters except letters
def clean_text(docs):
    cleaned_docs = []
    for doc in docs:
        cleaned_docs.append(' '.join([lemmatizer.lemmatize(word.lower())
                                        for word in doc.split()
                                        if letters_only(word)
                                        and word not in all_names]))
    return cleaned_docs
```

```python
cleaned_emails = clean_text(emails)
term_docs = cv.fit_transform(cleaned_emails)  #a sparse matrix termed_docs is created from the cleaned email data
print(term_docs [0])
```

## Define Functions and Paramters for training and testing data

```python
#Feature mapping is a function which maps the data vector to a feature space. 
feature_mapping = cv.vocabulary
feature_names = cv.get_feature_names()
```

```python
"""
    To generate a dictionary which has count has index/key and data from labels array as value,    
    Arguments: An array,labels
    Return: A dictionary, label_index     
"""
def get_label_index(labels):
    from collections import defaultdict
    label_index = defaultdict(list)
    for index, label in enumerate(labels):
        label_index[label].append(index)
    return label_index
```

```python
""" 
    To compute the prior of the label_index or any training subset,
    the following function get_prior is used. 
    
    Arguments: Dictionary with train indices, label_index
    Return: prior
"""
def get_prior(label_index):
    prior = {label: len(index) for label, index in label_index.items()}
    total_count = sum(prior.values())
    for label in prior:
        prior[label] /= float(total_count)
    return prior
```

```python
""" 
    To compute the likelihood of a training sample, 
    the following function get_likelihood is used.
    
    Arguments: A sparse matrix, term_document_matrix,
               A dictionary with train indices, label_index,
               Smoothing parameter, smoothing
    Return: dictionary,corresponding conditional probability P(feature|class) vector as value - likelihood
"""
def get_likelihood(term_document_matrix, label_index, smoothing=0):
    likelihood = {}
    for label, index in label_index.items():
        likelihood[label] = term_document_matrix[index, :].sum(axis=0) + smoothing
        likelihood[label] = np.asarray(likelihood[label])[0]
        total_count = likelihood[label].sum()
        likelihood[label] = likelihood[label] / float(total_count)
    return likelihood
```

```python
""" 
    To compute the posterior of testing samples,
    the following function get_posterior is used
    Arguments: A sparse matrix, term_document_matrix 
               prior 
               likelihood 
    Return: dictionary,corresponding posterior as value
"""
def get_posterior(term_document_matrix, prior, likelihood):
    num_docs = term_document_matrix.shape[0]
    posteriors = []
    for i in range(num_docs):
        # posterior is proportional to prior * likelihood
        # = exp(log(prior * likelihood))
        # = exp(log(prior) + log(likelihood))
        posterior = {key: np.log(prior_label) for key, prior_label in prior.items()}
        for label, likelihood_label in likelihood.items():
            term_document_vector = term_document_matrix.getrow(i)
            counts = term_document_vector.data
            indices = term_document_vector.indices
            for count, index in zip(counts, indices):
                posterior[label] += np.log(likelihood_label[index]) * count
        # exp(-1000):exp(-999) will cause zero division error,
        # however it equates to exp(0):exp(1)
        min_log_posterior = min(posterior.values())
        for label in posterior:
            try:
                posterior[label] = np.exp(posterior[label] - min_log_posterior)
            except:
                # if one's log value is excessively large, assign it infinity
                posterior[label] = float('inf')
        # normalize so that all sums up to 1
        sum_posterior = sum(posterior.values())
        for label in posterior:
            if posterior[label] == float('inf'):
                posterior[label] = 1.0
            else:
                posterior[label] /= sum_posterior
        posteriors.append(posterior.copy())
    return posteriors
```

## Split the data into folds using StratifiedKFold and carry out final training and testing using MultiNomialNB 

```python
from sklearn.model_selection import StratifiedKFold
k = 10
k_fold = StratifiedKFold(n_splits=k)
```

```python
# convert to numpy array for more efficient slicing
cleaned_emails_np = np.array(cleaned_emails)
labels_np = np.array(labels)
```

```python
"""Array max_feature_option, smoothing_factor_option and fit_prior_option are initialized
to compute efficiency of model at various thresholds """
max_features_option = [2000, 4000, 8000]
smoothing_factor_option = [0.5, 1.0, 1.5, 2.0]
fit_prior_option = [True, False]
auc_record = {}         #auc_record dictionary is created to save the score generated
```

```python
for train_indices, test_indices in k_fold.split(cleaned_emails, labels):
    X_train, X_test = cleaned_emails_np[train_indices], cleaned_emails_np[test_indices]   
    Y_train, Y_test = labels_np[train_indices], labels_np[test_indices]  
    for max_features in max_features_option:     
        if max_features not in auc_record:      
            auc_record[max_features] = {}
        cv = CountVectorizer(stop_words="english", max_features=max_features)
        term_docs_train = cv.fit_transform(X_train) 
        term_docs_test = cv.transform(X_test)           
        for smoothing_factor in smoothing_factor_option:   
            if smoothing_factor not in auc_record[max_features]:
                auc_record[max_features][smoothing_factor] = {}
            for fit_prior in fit_prior_option:     
                classifier = MultinomialNB(alpha=smoothing_factor, fit_prior=fit_prior) 
                classifier.fit(term_docs_train, Y_train) #The classifier is loaded with term_docs_train and Y_train data
                prediction_prob = classifier.predict_proba(term_docs_test) 
                pos_prob = prediction_prob[:, 1] 
                auc = roc_auc_score(Y_test, pos_prob) #The roc_auc_score or efficiency of classified model is computed
                auc_record[max_features][smoothing_factor][fit_prior] \
                    = auc + auc_record[max_features][smoothing_factor].get(fit_prior, 0.0) #the roc_auc_score is stored 
```
